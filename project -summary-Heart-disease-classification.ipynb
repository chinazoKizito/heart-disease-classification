#### Comparing Model Scores

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()



model_names =  ["Decision-Tree", "RandomForestCLF", "LogisticRegression", "KNN", 
                "XgboostCLF", "SVC(Kernel=Linear)", "SVC(Kernel=RBF)", "SVC(Kernel=Sigmoid)",
                "SVC(Kernel=poly)", "MultinomialNB"]

cross_val = [0.7649562, 0.8249462, 0.81161219, 0.8116129, 0.7981720, 0.82817204, 0.82827956, 0.8347311,
             0.8086021, 0.78537634]

cross_val_scores = []
for n in cross_val:
    a = np.round(n, 2)
    cross_val_scores.append(a)
    


df2 = pd.DataFrame({"Model_names": model_names, "Cross_val_score": cross_val_scores})
df2.sort_values("Cross_val_score", ascending=False, inplace=True)
    
fig, ax = plt.subplots(figsize=(8,6))
ax.barh(df2["Model_names"], df2["Cross_val_score"], color=['purple'])
plt.title(" Models Cross Validation Scores")
plt.gca().invert_yaxis()
plt.show()
```

```python
randomized_accuracy = [0, 84.21, 86.84, 86.84, 82.89, 0, 0, 0, 0, 0]
randomized_precision = [0, 82.22, 86.65, 87.80, 83.33, 0, 0, 0, 0, 0]
randomized_recall = [0, 90.24, 90.24, 87.80, 85.37, 0, 0, 0, 0, 0]
randomized_F1score = [0, 86.05, 88.10, 87.80, 84.34, 0, 0, 0, 0, 0]



df2 = pd.DataFrame({"Model_names": model_names, "randomized_accuracy": randomized_accuracy, 
                    "randomized_precision": randomized_precision, "randomized_recall": randomized_recall,
                    "randomized_F1_score": randomized_F1score})


a = df2.sort_values("randomized_accuracy", ascending=False)
b = df2.sort_values("randomized_precision", ascending=False)
c = df2.sort_values("randomized_recall", ascending=False)
d = df2.sort_values("randomized_F1_score", ascending=False)

    
fig, ax = plt.subplots(figsize=(8,6))
ax.barh(a["Model_names"], a['randomized_accuracy'], color=['blue'])
ax.set(title="Accuracy")
plt.gca().invert_yaxis()
plt.suptitle("RandomizedSearchCV Scores")
plt.show();

fig, ax = plt.subplots(figsize=(8,6))
ax.barh(b["Model_names"], b['randomized_precision'], color=['orange'])
ax.set(title="Precision")
plt.gca().invert_yaxis()
plt.show();

fig, ax = plt.subplots(figsize=(8,6))
ax.barh(c["Model_names"], c['randomized_recall'], color=['green'])
ax.set(title="Recall")
plt.gca().invert_yaxis()
plt.show();

fig, ax = plt.subplots(figsize=(8,6))
ax.barh(d["Model_names"], d['randomized_F1_score'], color=['red'])
ax.set(title="F1-score");
plt.gca().invert_yaxis()
plt.show()
```

```python
grid_accuracy = [0, 84.21, 86.84, 86.84, 84.21, 87.0, 89.0, 88.0, 88.0, 0]
grid_precision = [0, 82.22, 86.65, 87.80, 83.72, 84.0, 88.0, 88.0, 88.0, 0]
grid_recall = [0, 90.24, 90.24, 87.88, 87.80, 93.0, 93.0, 90.0, 90.0, 0]
grid_F1score = [0, 86.05, 88.10, 87.80, 85.71, 88.0, 90.0, 89.0, 89.0, 0]



df2 = pd.DataFrame({"Model_names": model_names, "Grid_accuracy": grid_accuracy, 
                    "Grid_precision": grid_precision, "Grid_recall": grid_recall,
                    "Grid_F1_score": grid_F1score})


a = df2.sort_values("Grid_accuracy", ascending=False)
b = df2.sort_values("Grid_precision", ascending=False)
c = df2.sort_values("Grid_recall", ascending=False)
d = df2.sort_values("Grid_F1_score", ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(d["Model_names"], d['Grid_F1_score'], color=['red'])
ax.set(title="F1-score")
plt.suptitle("GridSearchCV Scores")
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(c["Model_names"], c['Grid_recall'], color=['green'])
ax.set(title="Recall")
plt.suptitle("GridSearchCV Scores")
plt.xticks(rotation=90)
plt.show();
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(b["Model_names"], b['Grid_precision'], color=['orange'])
ax.set(title="Precision")
plt.suptitle("GridSearchCV Scores")
plt.xticks(rotation=90)
plt.show();
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(a["Model_names"], a['Grid_accuracy'], color=['blue'])
ax.set(title="Accuracy")
plt.xticks(rotation=90)
plt.suptitle("GridSearchCV Scores")
plt.show();
```
